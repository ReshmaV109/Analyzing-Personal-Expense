```python
from faker import Faker
import random
import pandas as pd
```

```python
fake=Faker()
```

```python
Categories=['Groceries', 'Subscriptions', 'Stationery', 'Gold investment', 'Fruits and vegetables', 'Transportation']
Payment_mode=['UPI', 'Cash', 'Netbanking', 'Credit card', 'Debit card']
```

```python
expense = {
    'Date': fake.date_this_month(),
    'Category': random.choice(Categories),
    'Payment_modes': random.choice(Payment_mode),
    'Description': fake.sentence(),
    'Amount': round(random.uniform(10,1000), 2)
}

print(expense)
```

```python
def gen_Exp_data(num_entries =150):
    data=[]
    for _ in range(num_entries):
        expense = {
    'Date': fake.date_this_month(),
    'Category': random.choice(Categories),
    'Payment_modes': random.choice(Payment_mode),
    'Description': fake.sentence(),
    'Amount': round(random.uniform(10,1000), 2)
     }
        data.append(expense)
    return pd.DataFrame(data)

exp_data=gen_Exp_data(150)
exp_data
```

```python
# Group by category and calculate total amount spent per category
category_total = df.groupby('Category')['Amount'].sum().reset_index()

# Calculate the total spending
total_spending = category_total['Amount'].sum()

# Calculate the percentage contribution of each category
category_total['Percentage'] = (category_total['Amount'] / total_spending) * 100

# Plot a pie chart
plt.figure(figsize=(8, 6))
plt.pie(
    category_total['Percentage'], 
    labels=category_total['Category'], 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=sns.color_palette('Set3', len(category_total))
)

# Adding a title
plt.title('Percentage Contribution of Each Category to Total Spending')

# Show the plot
plt.axis('equal')  # Equal aspect ratio ensures the pie chart is circular
plt.show()
```
